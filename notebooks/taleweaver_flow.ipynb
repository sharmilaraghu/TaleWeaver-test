# TaleWeaver вҖ” Code Flow Walkthrough

This notebook traces exactly what happens in a session вҖ” no UI needed.

## Full Request Flow

```
BEFORE SESSION (StoryScreen mounts)
  POST /api/story-plan   вҶ’ ADK LlmAgent вҶ’ 4-beat story arc + opening_text hint      вҶҗ NEW (google.adk)
  POST /api/story-opening вҶ’ Flash Lite (opening text) + Image model (first image)

SESSION START (child clicks "Begin")
  WebSocket /ws/story вҶ’ proxy.py вҶ’ Gemini Live API (voice, continues from opening)
    вҶі generate_illustration tool вҶ’ POST /api/image (skip_extraction) вҶ’ Image model
    вҶі showChoice tool вҶ’ ChoiceOverlay in browser
    вҶі awardBadge tool вҶ’ BadgePopup in browser

AFTER SESSION ENDS (child taps "рҹ“– See our story!")
  POST /api/story-recap вҶ’ Gemini (TEXT+IMAGE modalities) вҶ’ interleaved storybook    вҶҗ NEW (interleaved output)
```

## Step 3 вҖ” WebSocket Proxy Flow

After setup, `proxy.py` runs two async tasks in parallel:

```
Browser  в”Җв”Җ(audio chunks)в”Җв”Җв–¶  proxy_browser_to_gemini  в”Җв”Җв–¶  Gemini Live
Browser  в—Җв”Җв”Җ(audio + text)в”Җв”Җ  proxy_gemini_to_browser  в—Җв”Җв”Җ  Gemini Live
```

The browser sends **raw 16kHz PCM audio** (as base64 JSON). Gemini responds with **24kHz PCM audio** plus text transcripts.

Below is the exact shape of what the browser sends every ~100ms:

````python
# Shape of audio message from browser вҶ’ Gemini (the proxy forwards this as-is)
audio_message_shape = {
    'realtime_input': {
        'media_chunks': [
            {
                'mime_type': 'audio/pcm;rate=16000',
                'data': '<base64 encoded 16kHz PCM audio chunk>'
            }
        ]
    }
}

# Shape of response from Gemini вҶ’ browser
gemini_response_shape = {
    'serverContent': {
        'modelTurn': {
            'parts': [
                {'inlineData': {'mimeType': 'audio/pcm;rate=24000', 'data': '<base64 PCM audio>'}}
            ]
        },
        'outputTranscription': {'text': 'Once upon a time...'},  # what Gemini said
        'inputTranscription': {'text': 'Tell me more!'},          # what the child said
        'turnComplete': True,
    }
}

# Tool call shape (when Gemini decides to award a badge or show choices)
tool_call_shape = {
    'toolCall': {
        'functionCalls': [
            {
                'id': 'call_abc123',
                'name': 'awardBadge',
                'args': {'emoji': 'вӯҗ', 'name': 'Story Spark', 'reason': 'Suggested a dragon character'}
            }
        ]
    }
}

print('audio_message  вҶ’', json.dumps(audio_message_shape, indent=2)[:200], '...')
print('\ntool_call      вҶ’', json.dumps(tool_call_shape, indent=2))
````

## Step 4b (NEW) вҖ” `generate_illustration` Tool Call Flow

During the story, Gemini Live decides *when* to call `generate_illustration` at visually rich moments. The frontend handles it in `useLiveAPI.ts`:

```
Gemini calls generate_illustration({ scene_description: "..." })
  вҶ’ useLiveAPI immediately sends toolResponse (Gemini continues narrating вҖ” NOT blocked)
  вҶ’ forceImageGeneration(description) called:
      вҶ’ skip_extraction: true (Gemini's own description is already painter-ready)
      вҶ’ bypasses rate-limit check (tool images always fire)
      вҶ’ POST /api/image вҶ’ image gen вҶ’ StorySceneGrid
```

The `turnComplete` fallback (`triggerImageGeneration`) still runs at the user-configured interval for dialogue-heavy turns where Gemini doesn't call the tool.

````python
# Demonstrate skip_extraction вҖ” Gemini writes the scene description, no Flash Lite needed
from image_gen import ImageRequest, generate_scene_image

# Tool-triggered: scene_description is Gemini's own painter-ready text
tool_request = ImageRequest(
    scene_description="Wizard Wally standing on a silver moon crater, arms outstretched, conjuring a spiral of glowing blue star-sparks into the vast starry void above him, his purple robes billowing in zero gravity.",
    story_context="",
    image_style=wizard.image_style,
    session_id='notebook-tool-test',
    skip_extraction=True,         # вҶҗ key: bypass Flash Lite, use description as-is
    previous_image_data='',
    previous_image_mime_type='image/png',
    previous_scene_description='',
)

tool_result = await generate_scene_image(tool_request)
print(f'Scene used directly: {tool_result.scene_description}')
display(Image(data=base64.b64decode(tool_result.image_data)))
````

## Step 8 вҖ” Story Planner: ADK `LlmAgent`  вҶҗ NEW

Before a session starts, the **Story Planner** generates a structured story outline using Google ADK. This is TaleWeaver's explicit ADK layer вҖ” a `google.adk.agents.LlmAgent` backed by `gemini-2.0-flash`.

**File:** `backend/story_planner.py` В· **Endpoint:** `POST /api/story-plan`

### How it works

```
StoryScreen mounts
        в”Ӯ
        в–ј
  POST /api/story-plan  { character_id, theme }
        в”Ӯ
        в–ј
  google.adk.agents.LlmAgent  (story_planner)
    model: gemini-2.0-flash
    InMemorySessionService  вҶҗ  one session per request (plan-{character}-{hex})
    Runner.run_async()      вҶҗ  async generator, collect final response event
        в”Ӯ
        в–ј
  StoryPlan {
    setting:  "one sentence вҖ” where the story takes place"
    hero:     "child protagonist's opening situation"
    beats:    ["3вҖ“4 short plot beat strings, вүӨ10 words each"]
    ending:   "warm resolution sentence"
  }
        в”Ӯ
        в–ј
  opening_text hint  "[Story plan] Setting: ... | Start: ... | Moments: ... | Resolution: ..."
        в”Ӯ
        в–ј
  Passed as opening_text in the WebSocket init message
  вҶ’ proxy.py appends it to the "Begin!" trigger
  вҶ’ Gemini Live character has a creative roadmap from the first word
```

### ADK components used

| Component | Import | Role |
|---|---|---|
| `LlmAgent` | `google.adk.agents` | Wraps a Gemini model with a structured instruction |
| `Runner` | `google.adk.runners` | Executes the agent against a session |
| `InMemorySessionService` | `google.adk.sessions` | In-process session state (no DB needed) |
| `genai_types.Content` | `google.genai.types` | Structures the user message for the runner |

> Requires GCP credentials

````python
from story_planner import generate_story_plan, StoryPlanRequest

# The ADK LlmAgent generates a 4-beat story arc before the session starts.
# The returned opening_text can be passed in the WebSocket init message so the
# character has a creative roadmap from the very first word.

plan_request = StoryPlanRequest(
    character_id='wizard',
    theme='space adventure',
)

plan_result = await generate_story_plan(plan_request)

print('=== STORY PLAN (from ADK LlmAgent) ===')
print()
print(f'Setting:  {plan_result.plan.setting}')
print(f'Hero:     {plan_result.plan.hero}')
print()
print('Plot beats:')
for i, beat in enumerate(plan_result.plan.beats, 1):
    print(f'  {i}. {beat}')
print()
print(f'Ending:   {plan_result.plan.ending}')
print()
print('=== OPENING TEXT (injected into WebSocket init) ===')
print(plan_result.opening_text)
````

## Step 9 вҖ” Story Recap: Interleaved TEXT + IMAGE  вҶҗ NEW

After a story ends, the child taps **"рҹ“– See our story!"** to generate a full illustrated storybook. This uses **Gemini's native interleaved text+image output** вҖ” the Creative Storyteller category's mandatory technical requirement.

**Files:** `backend/image_gen.py` (endpoint) В· `frontend/src/components/StoryRecapModal.tsx` (UI)

### How it works

```
scene_descriptions[]          вҶҗ collected from generate_illustration tool calls
        в”Ӯ                        (stored in useStoryImages.ts, passed to modal on session end)
        в–ј
  gemini-2.0-flash-preview-image-generation
  GenerateContentConfig(response_modalities=["TEXT", "IMAGE"])
        в”Ӯ
        в”Ӯ  вҶҗ single model call, 90s timeout
        в–ј
  response.candidates[0].content.parts:
    Part 0 вҶ’ text  "Once upon a time, Wizard Wally gazed upon the stars..."
    Part 1 вҶ’ image  [PNG bytes of opening scene]
    Part 2 вҶ’ text  "But then, a tiny comet came zooming by..."
    Part 3 вҶ’ image  [PNG bytes of comet scene]
    ...
        в”Ӯ
        в–ј
  pages: [{ type:"text", content }, { type:"image", image_data, mime_type }, ...]
        в”Ӯ
        в–ј
  StoryRecapModal renders alternating <p> and <img> in a scrollable storybook layout
```

**Key detail:** `response_modalities=["TEXT", "IMAGE"]` tells the model to produce a single interleaved stream of text and images вҖ” it is NOT two separate calls. The model natively decides when to place each illustration relative to the narration.

> Requires GCP credentials + `IMAGE_MODEL` set to a Gemini model (not Imagen вҖ” Imagen doesn't support TEXT output)

````python
import base64
from IPython.display import Image, display
from image_gen import generate_story_recap, StoryRecapRequest

# Mock scenes вҖ” in a real session these come from generate_illustration tool calls,
# tracked in useStoryImages.ts and passed to StoryRecapModal when the story ends.
mock_scenes = [
    "Wizard Wally standing on a silver moon crater, arms outstretched, conjuring a spiral of glowing blue star-sparks into the vast starry void",
    "A tiny comet zooming past with a family of space mice riding on its glowing tail, waving hello",
    "Wally and the space mice sharing a warm moon-fruit picnic on the lunar surface, stars twinkling above",
]

recap_request = StoryRecapRequest(
    character_name='Wizard Wally',
    image_style=wizard.image_style,
    scene_descriptions=mock_scenes,
)

result = await generate_story_recap(recap_request)

print(f'Generated {len(result.pages)} storybook pages:')
print()
for i, page in enumerate(result.pages):
    if page.type == 'text':
        print(f'--- Narration ---')
        print(page.content)
        print()
    else:
        print(f'--- Illustration ({page.mime_type}) ---')
        display(Image(data=base64.b64decode(page.image_data)))
````
